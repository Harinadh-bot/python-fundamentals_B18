```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = pd.read_csv('countries (2) (1).csv')
x
```

```python
req: to do comparison analysis between two countries
```

```python
afg = x[x.country == 'Afghanistan']
```

```python
afg
```

```python
zim = x[x.country == 'Zimbabwe']
zim
```

```python
plt.plot(afg.year,afg.population)
plt.plot(zim.year,zim.population)
plt.xlabel('years')
plt.ylabel('population')
plt.legend(['afg','zim'])
plt.title('populational analysis')
plt.show()
```
