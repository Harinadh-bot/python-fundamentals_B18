```python
sales data
revenue v/s sales
x axis ---> sales
y axis ---> revenue
title of the graph ---> performance calculation of sales
```

```python
x = [1,2,3]
y = [1,4,9]
plt.title('performance calculation of sales')
plt.xlabel('sales')
plt.ylabel('Revenue')
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [0,5,10]
plt.title('performance calculation of sales')
plt.xlabel('sales')
plt.ylabel('Revenue')
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['this is x and y', 'this is x and z'])
plt.show()
```
